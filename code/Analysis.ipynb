```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\smith\Data_analysis\data\Data2.csv")  
df = df.drop("Unnamed: 0", axis=1)
df
```

```python
df.describe()
```

```python
df.var()
df["avg_d_kbps"].cov(df["avg_u_kbps"])
```

```python
#calculate the skewness adn kurtosis to evaluate the tail of the data and also if its positive or negatively skewed.

Continuous_columns = df.drop(["net_type_fixed", "net_type_mobile"] , axis = 1)

skewness_and_kurtois = Continuous_columns.agg(["skew", "kurtosis"]).transpose()
skewness_and_kurtois
```

```python
#plots each continuous data to show skewness
columns = Continuous_columns.columns.tolist()
for col in columns:
    sns.histplot(Continuous_columns[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
#normalise the data as its right skewed.
Continuous_columns_log = Continuous_columns.apply(np.log)
Continuous_columns_log
```

```python
cross_tab = pd.crosstab(df["net_type_fixed"], df["net_type_mobile"])
cross_tab
```

```python
#plotting each distrubtion after log

columns = Continuous_columns_log.columns.tolist()
for col in columns:
    sns.histplot(Continuous_columns_log[col], kde=True)
    plt.title(col)
    plt.show()
```

```python
#correlation matirx , when ran in my tensorflow enviorement it doesnt show all values
correlation_matrix = Continuous_columns_log.corr(method="pearson")
sns.heatmap(correlation_matrix, annot=True)
plt.show()
```

```python
#from the matrix i choose 3 pairs of features that correlate well strongly positive or negative:

plt.scatter(Continuous_columns_log["avg_u_kbps"], Continuous_columns_log["avg_d_kbps"])
plt.title("average download to upload kbps")
plt.show()

plt.scatter(Continuous_columns_log["avg_lat_down_ms"], Continuous_columns_log["avg_lat_up_ms"])
plt.title("average download to upload latency")
plt.show()

plt.scatter(Continuous_columns_log["avg_lat_up_ms"], Continuous_columns_log["avg_u_kbps"])
plt.title("average download to upload latency")
plt.show()
```
